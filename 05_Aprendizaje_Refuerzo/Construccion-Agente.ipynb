```python
import random
```

```python
class Environment:
  def __init__(self):
    self.pasos = 10
    self.flag = ["<- izq", "der ->"]
    self.steps = self.pasos                   # iteraciones
    self.board = [False for _ in range(5)]    # laberinto de 5 casillas se inicializan con False
    #print(self.board)
    self.board[4] = True                      # marca la casilla objetivo como True

  def reset(self):                            # Reinicio del ambiente
    self.steps = self.pasos

  def start(self) -> int:                     # Información sobre el límite inicial o izquierdo del ambiente
    return 0

  def end(self) -> int:                       # Información sobre el límite final o derecho del ambiente
    return len(self.board) - 1

  def is_done(self) -> bool:                  # Revisión si ya se acabo el episodio
    return self.steps == 0

  def action(self, state) -> bool:            # Definición del método del ambiente
    self.steps -= 1                           # Reducción de la cantidad de iteraciones disponibles
    print(f"Estado: {state} - Steps: {self.steps}")
    if self.is_done():                        # Revisión si ya acabaron las iteraciones
      print("-- Fin del episodio -- :( ")
    return self.board[state]
```

```python
class Agent:
  def __init__(self):
    self.state = 0                            # Variable para almacenar el estado actual
    self.actions = [0, 1]                     # Variable para almacenar las acciones

  def reset(self):                            # Reinicio del agente
    self.state = 0

  def forward(self, limit):                   # Movimiento hacia la derecha
    self.state = min(self.state + 1, limit)

  def backward(self, lowlimit):               # Movimiento hacia la izquierda
    self.state = max(self.state - 1, lowlimit)

  def action(self, env: Environment):         # Función para escoger la acción
    action = random.choice(self.actions)      # Escoger la acción de manera aleatoria entre 0 y 1
    if action:
      self.forward(env.end())                 # Información sobre el límite final o derecho del ambiente a través de env.end()
      #mov = 1
    else:
      self.backward(env.start())              # Información sobre el límite inicial o izquierdo del ambiente a través de env.start()
      #mov = 0
    mov = action
    print(env.flag[mov])
    return env.action(self.state)
```

```python
def main():                                   # Función para crear un loop entre el ambiente y el agente - ejecución del agente donde el agente recibe info del estado y su recompensa
  episodes = 5
  a = Agent()
  e = Environment()
  for i in range(0, episodes):
    print(f"==================== \n EPISODIO: {i+1}")
    count = 0
    while not e.is_done():
      count += a.action(e)
    print(f"{count} éxitos - {e.pasos - count} fracasos")
    print(f"ESTADO FINAL: {a.state}")
    a.reset()
    e.reset()

main()
```
