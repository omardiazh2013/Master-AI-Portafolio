```python
#Definición del ambiente de motores

#Librerias requeridas
import random
#raise NotImplementedError

#Definición de la clase principal
class MDP:
    def __init__(self, table : list[list[tuple]], dimensions:tuple[int, int], initial_state:int):
        self.table = table
        self.ncols = dimensions[1]
        self.nrows = dimensions[0]
        self.initial_state = initial_state
        self.state = initial_state

        # Creacion de atributo 'transitions' y 'rewards'
        self.transitions = []
        self.rewards = []
        tuplas = [t for fila in table for celda in fila if celda is not None for t in celda]
        for t in tuplas:
            self.transitions.append((t[0], t[1]))
            self.rewards.append((t[2], t[1]))


    def get_current_state(self) -> int:
        return self.state

    def get_possible_actions(self, state:int) -> list[str]:
        actions = []
        if state == 2:
            return actions
        else:
            actions = ['slow', 'fast']
            return actions

    def do_action(self, action:str) -> tuple[float,int]:
        if action == 'slow' and self.state == 0:
            return self.rewards[0]
        elif action == 'fast' and self.state == 0:
            i = random.choice([1, 2])
            self.state = self.rewards[i][1]
            return self.rewards[i]
        elif action == 'slow' and self.state == 1:
            i = random.choice([3, 4])
            self.state = self.rewards[i][1]
            return self.rewards[i]
        elif action == 'fast' and self.state == 1:
            self.state = self.rewards[5][1]
            return self.rewards[5]
        elif self.state == 2:
            return ('NA', 2)

    def reset(self):
        self.state = self.initial_state

    def is_terminal(self) -> bool:
        if self.state == 2:
            return True
        else:
            return False
```

```python
import random

table = [[[(1,0,1)], [(0.5, 0, 2), (0.5,1,2)]],
         [[(0.5, 0, 1), (0.5,1,1)], [(1, 2, -10)]],
         [None, None]]

env = MDP(table, (3,2), 0)

done = False
total_reward = 0
steps = 0
while not done:
    old_state = env.get_current_state()
    action = random.choice(env.get_possible_actions(old_state))
    reward, new_state = env.do_action(action)
    total_reward += reward
    print(f"Ejecutando la acción {action} desde el estado {old_state} al estado {new_state}")
    steps += 1
    done = env.is_terminal()
print(f"Ejecuto {steps} pasos y obtuvo una recompensa de {total_reward}")
```
