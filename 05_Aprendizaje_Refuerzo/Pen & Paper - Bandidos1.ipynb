```python
import random
lista_primos = [2, 3, 5, 7, 11, 13, 17, 19]
```

```python
def camino(x):
    factores = []
    for p in lista_primos:
        while x % p == 0:
            factores.append(p)
            x //= p
    return len(factores)
```

```python
# Definicion de la clase Bandit
class Bandit:
    def __init__(self, num_arms=4):
        self.num_arms = num_arms
        self.rewards = [0 for _ in range(num_arms)]
        self.occurrences = [0 for _ in range(num_arms)]
        self.cumulative_rewards = [0 for _ in range(num_arms)]
        self.list_rewards = [18, 8, 16, 12]

    def expected_reward(self, arm:int, hora:int) -> float:
        self.occurrences[arm] += 1

        # Regla principal
        if hora % 4 == 0:
            reward = 0
        else:
            reward = self.list_rewards[arm]

        n = self.occurrences[arm]
        # self.cumulative_rewards[arm] += reward
        self.rewards[arm] = self.rewards[arm] + (reward - self.rewards[arm]) / n
        return reward

    def choose_arm(self, hora:int) -> int:
        return min(camino(hora) - 1, self.num_arms - 1)

    def choose_arm_random(self, hora:int, epsilon=0.5) -> int:
        if random.random() < epsilon:
            return self.choose_arm(hora)
        else:
            return random.randint(0, self.num_arms - 1)
```

```python
class Bandit(Bandit):
  def run(self):
    for hora in range(7, 21):
        arm = self.choose_arm(hora)
        self.expected_reward(arm, hora)
    return self.rewards
```

```python
class Bandit1(Bandit):
    def run(self):
        for hora in range(7, 21):
            arm = self.choose_arm_random(hora)
            self.expected_reward(arm, hora)
        return self.rewards
```

```python
b = Bandit()
b.run()
```
