```python
# Define los estados
HIGH = "high"
LOW = "low"

# Define las acciones
SEARCH = "search"
WAIT = "wait"
RECHARGE = "recharge"

# Define probabilidades (α, 1-α, β, 1-β)
ONE = "1"
ZERO = "0"

ALPHA = "alpha"
ONE_MINUS_ALPHA = f"{ONE}-{ALPHA}"

BETA = "beta"
ONE_MINUS_BETA = f"{ONE}-{BETA}"

# Define recompensas
R_SEARCH = "r_search"
R_MINUS_3 = "-3"
R_WAIT = "r_wait"
R_NONE = "-"
R_ZERO = "0"

def get_transition_table():
    ''' TODO: Completar la tabla de transición como lista de listas de strings.
    Cada lista deberá tener este formato: [estado, acción, estado_resultante, probabilidad, recompensa]
    '''
    transition_table = [
        [HIGH, SEARCH, HIGH, ALPHA, R_SEARCH],
        [HIGH, SEARCH, LOW, ONE_MINUS_ALPHA, R_SEARCH],
        [HIGH, WAIT, HIGH, ONE, R_WAIT],
        [LOW, SEARCH, LOW, BETA, R_SEARCH],
        [LOW, WAIT, LOW, ONE, R_WAIT],
        [LOW, RECHARGE, HIGH, ONE, R_ZERO],
        [LOW, RECHARGE, LOW, ZERO, R_NONE],
        [HIGH, WAIT, LOW, ZERO, R_NONE],
        [LOW, WAIT, HIGH, ZERO, R_NONE],
        [LOW, SEARCH, HIGH, ONE_MINUS_BETA, R_MINUS_3]
    ]
    # Devolver la tabla
    return transition_table
```
