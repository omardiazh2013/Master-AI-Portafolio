## Tokenización del corpus de texto en palabras
Defina el cuerpo de la función **tokenize** de modo que transforme el corpus de texto en una listado de tokens a nivel de palabras. Solo modifique el código en el lugar que se le indica. <br>
Recuerde que algunas de las librerías retornan listados de objetos, en estos casos deberá transformar la lista para que sus elementos sean cadenas de texto (tipo str).

```python
def tokenize (text):
    """
    Esta función transforma un texto en una lista de tokens a nivel de palabra.

    Args:
    - text (str): El texto a tokenizar.

    Returns:
    tokens (list[str]): Una lista con los tokens en string a nivel de palabra.
    """

    ## INICIO DE IMPLEMENTACIÓN (ÚNICAMENTE MODIFIQUE EL CÓDIGO AQUÍ)
    # your code here
    
    tokens = text.split()
    return tokens
    
    #raise NotImplementedError
    ## FIN DE IMPLEMENTACIÓN (A PARTIR DE ESTE PUNTO HACIA ABAJO NO DEBE REALIZAR NINGUNA MODIFICACIÓN)

tokens = tokenize(corpus)
tokens[:10]
```

## Eliminación de tokens de puntuación y transformación a minúsculas
Complete el cuerpo de la función **clean_punct** para normalizar las palabras eliminando los tokens pertenecientes a caracteres de puntuación, y transformado los tokens a minúsculas. 
Recuerde que algunas de las librerías retornan listados de objetos, en estos casos deberá transformar la lista para que sus elementos sean cadenas de texto (tipo str).

```python
def clean_punct(tokens):
    """
    Esta función transforma una lista de tokens eliminando los tokens de puntuación y transformandolos a minúsculas.

    Args:
    - tokens(list[str]): Los tokens a procesar.

    Returns:
    tokens (list[str]): Una lista con los tokens limpios, sin tokens de puntuación y en minúsculas. Un token debe estar compuesto únicamente por caracteres alfanúmericos.
    """

    ## INICIO DE IMPLEMENTACIÓN (ÚNICAMENTE MODIFIQUE EL CÓDIGO AQUÍ)
    # your code here
    tokens_limpios = [t.lower() for t in tokens if t.isalnum()]
    
    return tokens_limpios
    #raise NotImplementedError
    ## FIN DE IMPLEMENTACIÓN (A PARTIR DE ESTE PUNTO HACIA ABAJO NO DEBE REALIZAR NINGUNA MODIFICACIÓN)

tokens_sin_punct = clean_punct(tokens)
tokens_sin_punct[:10]
```

## Eliminación de palabras de parada
Defina el cuerpo de la función **remove_stopwords**, en donde debe eliminar las palabras de parada del idioma español.<br>
Recuerde que algunas de las librerías retornan listados de objetos, en estos casos deberá transformar la lista para que sus elementos sean cadenas de texto (tipo str).

```python
def remove_stopwords(tokens):
    """
    Esta función transforma una lista de tokens eliminando los tokens pertencientes a palabras de parada del idioma.
    
    Args:
    - tokens[str]): Lita de tokens.

    Returns:
    tokens (list[str]): Una lista con los tokens sin palabras de parada.
    """

    ## INICIO DE IMPLEMENTACIÓN (ÚNICAMENTE MODIFIQUE EL CÓDIGO AQUÍ)
    # your code here
    tokens_sin_stop = [t for t in tokens if t not in stop_words_es]
    return tokens_sin_stop
    #raise NotImplementedError
    ## FIN DE IMPLEMENTACIÓN (A PARTIR DE ESTE PUNTO HACIA ABAJO NO DEBE REALIZAR NINGUNA MODIFICACIÓN)

tokens_sin_stopwords = remove_stopwords(tokens_sin_punct)
tokens_sin_stopwords[:10]
```

## Lematización
Complete la definición de la función **lemmatize**, esta función debe transformar a su lema cada uno de los tokens.<br>

```python
def lemmatize(tokens):
    """
    Esta función transforma una lista de tokens a su respectivo lema en el idioma español.

    Args:
    - tokens(list[str]): Lista de tokens.

    Returns:
    tokens (list[str]): Una lista con los lemas de cada token.
    """

    ## INICIO DE IMPLEMENTACIÓN (ÚNICAMENTE MODIFIQUE EL CÓDIGO AQUÍ)
    # your code here
    text = " ".join(tokens)
    doc = nlp(text)
    tokens_lematizados = [t.lemma_ for t in doc]
    return tokens_lematizados
    #raise NotImplementedError
    ## FIN DE IMPLEMENTACIÓN (A PARTIR DE ESTE PUNTO HACIA ABAJO NO DEBE REALIZAR NINGUNA MODIFICACIÓN)

lemmas = lemmatize(tokens_sin_stopwords)
lemmas[:10]
```

```python
# Celda no modificable

inital_tokens = tokenize(corpus) # Texto tokenizado a nivel de palabras
processed_tokens = lemmas # Texto con las 4 etapas de procesamiento de texto ejecutadas

unique_tokens = len(set(inital_tokens))
unique_lemmas = len(set(processed_tokens))

print(f"Número de tokens únicos iniciales: {unique_tokens}")
print(f"Número de tokens únicos procesados: {unique_lemmas}")
```
