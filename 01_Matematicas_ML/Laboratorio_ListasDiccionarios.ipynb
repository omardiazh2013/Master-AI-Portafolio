
# **Introducción a listas y diccionarios**

##**1. ¿Qué son las listas?**

Las listas son una estructura de datos nativa de python que permite almacenar información de manera ordenada y sencilla. Una lista puede almacenar elementos de manera ordenada, donde sus elementos se pueden editar y repetir.

Las listas se pueden definir en python con un par de corchetes cuadrados que almacenan los datos. Una lista con los números de 1 a 5 puede ser creada de la siguiente manera:

```
lista = [1,2,3,4,5]
```
Para obtener la cantidad de elementos dentro de la lista se utiliza la función **len** de la siguiente manera:
```
len(lista)
```


**1.1. Indexación de listas**

Para acceder a un elemento dentro de la lista se debe "indexar" el elemento mediante su posición. Es importante recordar que la primera posición de una lista es la posición 0. Así, si se quiere extraer el tercer elemento de la lista definida anteriormente, se debe hacer lo siguiente:
```
lista[2]
```
Note que el tercer elemento corresponde a la posición 2 dentro de la lista debido a que el primer elemento está en la posición 0, el segundo elemento en la posición 1 y el tercer elemento en la posición 2.

Cuando se quiere acceder a distintos elementos dentro de la lista se debe hacer uso de dos puntos ":" para delimitar los elementos que se quieren extraer. Asuma que quiere extraer los elementos desde el segundo (índice 1) hasta el cuarto (índice 3) de la lista definida anteriormente, el código que realiza la operación mencionada es el siguiente:

```
lista[1:4]
```
Note que en este caso el último índice es 4 y no 3, esto se debe a que cuando se indexa una lista de esta manera, la indexación retorna los elementos desde el inicial hasta el final menos uno.

**1.2. Métodos sobre listas**

Algunos de los métodos más utilizados y sus respectivas acciones se definen a continuación:

**append:** Este método añade el elemento dado por parámetro al final de la lista.

**index:** Este método retorna la posición del elemento dado por parámetro, en caso de que el elemento dado no se encuentre en la lista, retorna None.

**pop:** Este método remueve el elemento en el índice dado por parámetro y lo devuelve.

**remove:** Este método remueve el elemento dado por parámetro.

**sort:** Este método devuelve la lista ordenada.

##**2. ¿Qué son los diccionarios?**

Los diccionarios son estructuras de datos nativas de python que permiten almacenar información mediante el uso de parejas llave - valor *(key,value)*. Dichos elementos pueden ser modificados pero sus llaves no pueden ser repetidas. A partir de python 3.7 los diccionarios son ordenados.

Los diccionarios en python se definen con corchetes que almacenan las parejas llave valor separadas por coma (,). Un diccionario con la ubicación del continente de los países Colombia, España y Canadá se define de la siguiente manera:

```
diccionario = {"Colombia":"Sudamérica","España":"Europa","Canadá":"Norteamérica"}
```
Para obtener la cantidad de parejas llave - valor dentro de un diccionario se puede utilizar la función *len* de la siguiente manera:
```
len(diccionario)
```


**2.1. Indexación de diccionarios**

Si se desea acceder a un valor dentro del diccionario se debe utilizar su respectiva llave entre corchetes cuadrados. Por ejemplo, para acceder al continente de España se realiza lo siguiente:

```
diccionario["España"]
```

**2.2. Métodos sobre diccionarios**

Algunos de los métodos más utilizados en diccionarios y su función se muestran a continuación:

**keys:** Este método retorna todas las llaves del diccionario en un formato iterable.

**values:** Este método retorna todos los valores del diccionario en un formato iterable.

**items:** Este método retorna todas las parejas llave-valor dentro de tuplas en un formato iterable.

**get:** Este método retorna el valor de la llave ingresada por parámetro, en caso de que no exista la llave dentro del diccionario retorna None.
