\begin{equation}

**c) Implemente un código en Python que reciba un vector de entrada 𝑥∈ℝ2 y un ángulo en radianes 𝜃, y devuelva el vector rotado. Además, este código debería graficar el vector original y el vector rotado (en colores diferentes).**

```python
import numpy as np
import matplotlib.pyplot as plt

def rotar_vector(x, theta):
    # Matriz de rotación
    R = np.array([[np.cos(theta), np.sin(theta)],
                  [-np.sin(theta), np.cos(theta)]])

    # Multiplicar la matriz por el vector
    x_rotado = np.dot(R, x)
    print(R)
    return x_rotado

def graficar_vectores(x, x_rotado):
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', linewidth=0.5)

    # Graficar el vector original en azul
    plt.quiver(0, 0, x[0], x[1], color='blue', angles='xy', scale_units='xy', scale=1, label='Original')
    # Graficar el vector rotado en rojo
    plt.quiver(0, 0, x_rotado[0], x_rotado[1], color='red', angles='xy', scale_units='xy', scale=1, label='Rotado')

    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.legend()
    plt.show()
```

Se presentan varios ejemplos, dado un vector [1, 1] con ángulos de 30, 45 y 90 grados

```python
x1 = np.array([1, 1])
theta = np.pi /6
x_rotado = rotar_vector(x1, theta)
graficar_vectores(x1, x_rotado)
#
```
