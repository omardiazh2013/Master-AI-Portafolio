```python
class Gridworld:
    def __init__(self, board : list[list[str]], dimensions: tuple[int,int]):
        self.grid = board
        self.ncols = dimensions[1]
        self.nrows = dimensions[0]

        # Búsqueda del estado inicial
        for i in range (self.nrows):
            for j in range (self.ncols):
                if self.grid[i][j] == 'S':
                    self.initial_state = (i,j)

        # Inicialización del estado
        self.state = self.initial_state

        # Imposicion de las recompensas en el ambiente
        for i in range (self.nrows):
            for j in range (self.ncols):
                if self.grid[i][j] == ' ':
                    self.grid[i][j] = 0
                elif self.grid[i][j] == '#':
                    self.grid[i][j] = None
                elif self.grid[i][j] == '-1':
                    self.grid[i][j] = int(-1)
                elif self.grid[i][j] == '+1':
                    self.grid[i][j] = int(1)

    def get_current_state(self) -> tuple[int,int]:
        return self.state

    def reset(self):
        self.state = self.initial_state

    def get_possible_actions(self, state:tuple[int,int]) -> list[str]:
        actions = []

        if self.grid[state[0]][state[1]] == None:
            return actions

        elif self.grid[state[0]][state[1]] == 1 or self.grid[state[0]][state[1]] == -1:
            actions.append('exit')
            return actions

        #elif state[0] > 0 and self.grid[state[0]-1][state[1]] != None:
        #    actions.append('up')
        #if state[1] < (self.ncols -1) and self.grid[state[0]][state[1]+1] != None:
        #    actions.append('right')
        #if state[0] < (self.nrows - 1) and self.grid[state[0]+1][state[1]] != None:
        #    actions.append('down')
        #if state[1] > 0 and self.grid[state[0]][state[1]-1] != None:
        #    actions.append('left')
        #return actions

        else:
          actions = ['up', 'right', 'down', 'left']
          return actions


    def do_action(self, action:str) -> tuple[float, tuple[int,int]]:
        # Movimiento hacia arriba
        if action == 'up':
            if self.state[0] == 0 or self.grid[self.state[0]-1][self.state[1]] == None:
                return self.grid[self.state[0]][self.state[1]], self.state
            else:
                self.state = (self.state[0]-1, self.state[1])
                return self.grid[self.state[0]][self.state[1]], self.state
        # Movimiento hacia la derecha
        elif action == 'right':
            if self.state[1] == (self.ncols - 1) or self.grid[self.state[0]][self.state[1]+1] == None:
                return self.grid[self.state[0]][self.state[1]], self.state
            else:
                self.state = (self.state[0], self.state[1]+1)
                return self.grid[self.state[0]][self.state[1]], self.state
        # Movimiento hacia abajo
        elif action == 'down':
            if self.state[0] == (self.nrows - 1) or self.grid[self.state[0]+1][self.state[1]] == None:
                return self.grid[self.state[0]][self.state[1]], self.state
            else:
                self.state = (self.state[0]+1, self.state[1])
                return self.grid[self.state[0]][self.state[1]], self.state
        # Movimiento hacia izquierda
        elif action == 'left':
            if self.state[1] == 0 or self.grid[self.state[0]][self.state[1]-1] == None:
                return self.grid[self.state[0]][self.state[1]], self.state
            else:
                self.state = (self.state[0], self.state[1]-1)
                return self.grid[self.state[0]][self.state[1]], self.state

    def is_terminal(self) -> bool:
        if self.grid[self.state[0]][self.state[1]] == 1 or self.grid[self.state[0]][self.state[1]] == -1:
            return True
        else:
            return False
```
