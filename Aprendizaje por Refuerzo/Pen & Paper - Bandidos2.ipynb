```python
import random
```

```python
def propina(cliente, pelicula) -> float:
  return min(cliente * pelicula / 2, 5)
```

```python
# Definicion de la clase Bandit
class Bandit:
    def __init__(self, num_movies=4):
        self.num_movies = num_movies
        self.rewards = [10.0 for _ in range(num_movies)]
        self.list_peliculas = [4, 2, 3, 3, 3, 1, 4, 2, 1, 2, 3, 3, 2, 2, 3, 4, 4, 4, 3, 1]
        self.list_criterio = [i not in [1,3,4,11,12,13,18] for i in range(1,21)]
        self.precio_base = 10

    def expected_reward(self, movie:int, cliente:int, alpha = 0.1) -> float:
        # Regla principal
        if self.list_criterio[cliente]:
            reward = self.precio_base + propina(cliente + 1, movie + 1)
        else:
            reward = 0

        self.rewards[movie] = self.rewards[movie] + alpha * (reward - self.rewards[movie])
        return reward

    def choose_movie(self, cliente:int) -> int:
        return self.list_peliculas[cliente] - 1

```

```python
class Bandit(Bandit):
  def run(self):
    for cliente in range(0, 20):
        peli = self.choose_movie(cliente)
        self.expected_reward(peli, cliente)
    return [round(r, 2) for r in self.rewards]
```

```python
b = Bandit()
b.run()
```
